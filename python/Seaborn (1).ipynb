seaborn is powerful library for visualization built on top of matplotlib that provide high level and attractive statistical graphics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head(1)
```

#histogram - data distribution

x - numerical(continious feature)

y - frequency - count

bins -

```python
plt.figure(figsize=(8,5))
sns.histplot(data = titanic , x = 'age', bins = 30 , kde= True)
plt.title("Age distribution")

#KDE - Kernal Density Estimator
```

#KDE PLOT

```python
sns.kdeplot(data = titanic,x='age', fill= True, color= 'purple')
```

#SCATTER PLOT

x - numerical - continious

y - numerical - continious

hue - always define categorical feature

to visualize the relation of two variable that is x incereses is y increses or not ?

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data = titanic, x ='age', y ='fare', hue= 'sex', palette= 'Set1')

plt.title("scatter plot of age and fare")
```

#Line plot - trends

x - numerical - continious

y - numerical continious

```python
plt.figure(figsize=(8,5))

sns.lineplot(data = titanic, x ='age', y='fare', hue ='sex', palette= 'viridis')

plt.title("Line plot")
```

#Bar plot

x - categorical
y - numerical

```python
plt.figure(figsize=(10,6))

sns.barplot(data = titanic, x= 'pclass', y='fare', hue='sex',  palette= 'rocket')

plt.title("Total fare by pclass")

plt.legend()
```

#COUNT PLOT

x - caetorical data count


```python
titanic.groupby('sex')['survived'].value_counts().unstack()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(data = titanic , x = 'sex', hue = 'survived',  palette= 'hls')
```

#BOXPLOT

```python
sns.boxplot(data=titanic, x ='age', y ='sex', hue = 'survived', palette= 'Set1')
```

#violin plot  - same as boxplot. define outliers

```python
sns.violinplot(data = titanic, x='age', y ='sex')
```

#Hatmap

```python
sns.heatmap(titanic.corr(numeric_only=True), cmap='coolwarm', annot= True)
```

#PAIR PLOT

```python
sns.pairplot(data = titanic)
```

```python
sns.swarmplot(data = titanic, y ='sex')
```
